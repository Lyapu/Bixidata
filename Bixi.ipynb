```python
#parse an xml file
def add_data_from_file(path, filename):
    #print path, filename
    tree = ET.parse(os.path.join(path, filename))
    time = filename[12:20]
    hour = filename[12:14]
    minute = filename[15:17]
    day = path[-2:]
    dt = datetime.datetime(2012, month=6, day=int(day), hour=int(hour), minute=int(minute))
    root = tree.getroot()
    i = len(df)
    for station in root:
        data = {}
        data["time"] = dt.time()
        data["date"] = dt.date()
        for child in station:
            data[child.tag] = child.text
        #print data
        df.loc[i] = data
        i += 1
```

```python
import math
def distance_from_center(x):
    long1 = -73.567224
    lat1 = 45.501898
    long2 = float(x["long"])
    lat2 = float(x["lat"])
    
    # Convert latitude and longitude to 
    # spherical coordinates in radians.
    degrees_to_radians = math.pi/180.0
    # phi = 90 - latitude
    phi1 = (90.0 - lat1)*degrees_to_radians
    phi2 = (90.0 - lat2)*degrees_to_radians
    # theta = longitude
    theta1 = long1*degrees_to_radians
    theta2 = long2*degrees_to_radians
    cos = (math.sin(phi1)*math.sin(phi2)*math.cos(theta1 - theta2) + 
           math.cos(phi1)*math.cos(phi2))
    arc = math.acos( cos )

    return arc*6373
```

```python
df["dist_from_center"] = df.apply(distance_from_center, axis=1)
```
